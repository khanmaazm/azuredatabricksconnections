```
# Databricks notebook source
import time
import urllib
import hmac
import hashlib
import base64
```

```

# COMMAND ----------

sb_name="aimindsfirsteventhub"
eh_name="custdetailevents"
sas_name="RootManageSharedAccessKey"
sas_value="9O5Q2ygao9JlOgeSFpei04scFMmPNO4HnTaD3A6XKXA="

```

```

# COMMAND ----------

uri = urllib.parse.quote_plus("https://{}.servicebus.windows.net/{}" \
                                  .format(sb_name, eh_name))

# COMMAND ----------

print(uri)
```

```
# COMMAND ----------

sas = sas_value.encode('utf-8')
expiry = str(int(time.time() + 200000000))
print (expiry)
string_to_sign = (uri + '\n' + expiry).encode('utf-8')
signed_hmac_sha256 = hmac.HMAC(sas, string_to_sign, hashlib.sha256)
signature = urllib.parse.quote(base64.b64encode(signed_hmac_sha256.digest()))
print('SharedAccessSignature sr={}&sig={}&se={}&skn={}' \
                 .format(uri, signature, expiry, sas_name))
print("https://", sb_name, ".servicebus.windows.net/", eh_name, "/messages")
```
